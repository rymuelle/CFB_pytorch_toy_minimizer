```python
import pandas as pd
import numpy as np
import requests
from time import sleep
```

```python
#get list of teams
header = {"accept": "application/json"}
teams = requests.get('https://api.collegefootballdata.com/teams', json=header).json()
```

```python
team_df = pd.DataFrame(teams)
team_df.head()
team_df.to_csv("cfb_api_data/teams.csv")
```

```python
first_year = 2000 
last_year = 2020
year_list = np.linspace(first_year,last_year,last_year-first_year+1,dtype=int)
recruiting_list = []
for year in year_list:
    recruiting_scores = requests.get("https://api.collegefootballdata.com/recruiting/teams?year=".format(year), json=header).json()
    recruiting_list.extend(recruiting_scores)
    sleep(2)
```

```python
#get games box scores
first_year = 2004 
last_year = 2020
year_list = np.linspace(first_year,last_year,last_year-first_year+1,dtype=int)
week_list = np.linspace(1,20,20,dtype=int)
header = {"accept": "application/json"}
game_boxscore_list = []
for year in year_list:
    #regular season
    print(year)
    for week in week_list:
        games_box = requests.get("https://api.collegefootballdata.com/games/teams?year={}&week={}&seasonType=regular".format(year,week), json=header).json()
        #pprint.pprint(games_box)
        if games_box == []: break
        game_boxscore_list.extend(games_box)
    #post season
    for week in week_list:
        games_box = requests.get("https://api.collegefootballdata.com/games/teams?year={}&week={}&seasonType=postseason".format(year,week), json=header).json()
        if games_box == []: break
        game_boxscore_list.extend(games_box)
    sleep(4)
    print(year, len(game_boxscore_list))
```

```python
game_box_df = pd.DataFrame(game_boxscore_list)
game_box_df.head()
game_box_df.to_csv("cfb_api_data/games_boxscore.csv")
```
